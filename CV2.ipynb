```python
#import library
import cv2 #Pythons image processing library 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
color_1=[255,0,0]
color_2=[0,255,0]
color_3=[0,0,255]
color_4=[125,125,125]
plt.imshow(np.array([[color_1,color_2],[color_3,color_4]]))
```

```python
#add a text of the color code of each pixel

color=[[color_1,color_2],[color_3,color_4]]
print (np.array(color).shape)
plt.imshow(np.array(color))

start_row=0
for row in color:
  start_col=-0.25
  for color in row:
    plt.text(start_row,start_col,str(color))
    start_col+=1
  start_row+=1
```

**Import** **Image**



*   Using cv2 library to import an image 
*   Perform few basic operations on the image 



```python
#read image
image=cv2.imread('/content/waynerooney.jpg')
print (type(image))
```

```python
image.shape
```

```python
plt.imshow(image)
```

```python
#imported images always follow BGR as the pixel color oder which is converted to RGB to maintain the actual image colors
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
# image in gray scale
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray',)
```

```python
#Resizing the image
resized=cv2.resize(image,(300,300))
plt.imshow(resized)
```

```python
#Maintaining the aspect ratio while resizing 
height=300
width=300
aspect= image.shape[1]/float(image.shape[0])
aspect
if (aspect>1):#landscape
  res=int(aspect*height)
  scaled=cv2.resize(image,(res,height))
if (aspect>1):#portrait
  res=int(width/aspect)
  scaled=cv2.resize(image,(width,res))
else:
  scaled=cv2.resize(image,(width,height))

plt.imshow(scaled,cmap='gray')
```

```python
scaled.shape
```

```python
#corping a section of the image 

def crop_center(img,cropx,cropy):
  y,x,c=img.shape
  startx=x//2-(cropx//2)
  starty=y//2-(cropy//2)
  return img[starty:starty+cropy,startx:startx+cropx]

cropped=crop_center(image,width,height)
plt.imshow(cropped,cmap='gray')
```

```python
#adding a texts and lines to the image 
cv2.putText(image,'MUFC Legend',(160,169),cv2.FONT_HERSHEY_SIMPLEX,5,(255,0,0),5)
cv2.line(image,(0,0,),(500,500),(0,0,0),20)
plt.imshow(image,cmap='gray')
```
